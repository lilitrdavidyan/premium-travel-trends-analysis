```python
print("Start Date Range: ", data['startDate'].min(), "to", data['startDate'].max())
print("End Date Range: ", data['endDate'].min(), "to", data['endDate'].max())
```

```python
plt.figure(figsize=(12, 6))

plt.hist(data['startDate'], bins=30, alpha=0.5, label='Start Date')
plt.hist(data['endDate'], bins=30, alpha=0.5, label='End Date')

plt.title('Distribution of Start and End Dates')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

```python
data['startDate'] = pd.to_datetime(data['startDate'])
data['endDate'] = pd.to_datetime(data['endDate'])
```

```python
plt.figure(figsize=(12, 6))

data.set_index('startDate').resample('M').size().plot(label='Start Date', linewidth=2)
data.set_index('endDate').resample('M').size().plot(label='End Date', linewidth=2)

plt.title('Number of departures over time')
plt.ylabel('Number of departures')
plt.xlabel('Time')
plt.legend()
plt.show()
```
