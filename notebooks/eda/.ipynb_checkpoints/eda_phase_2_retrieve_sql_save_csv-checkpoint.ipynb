```python
import pandas as pd
import mysql.connector
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
DB_HOST = os.getenv('DB_HOST')
DB_USER = os.getenv('DB_USER')
DB_PASSWORD = os.getenv('DB_PASSWORD')
DB_NAME = os.getenv('DB_NAME')
```

```python
DATABASE_URI = f'mysql+mysqlconnector://{DB_USER}:{DB_PASSWORD}@{DB_HOST}/{DB_NAME}'
engine = create_engine(DATABASE_URI)
```

```python
tours_query = "SELECT * FROM tours;"
tour_options_query = "SELECT * FROM tour_options;"
itineraries_query = "SELECT * FROM itineraries;"
locations_visited_query = "SELECT * FROM locations_visited;"
options_query = "SELECT * FROM options;"
departures_query = "SELECT * FROM departures;"
```

```python
tours_df = pd.read_sql(tours_query, engine)
tour_options_df = pd.read_sql(tour_options_query, engine)
itineraries_df = pd.read_sql(itineraries_query, engine)
locations_visited_df = pd.read_sql(locations_visited_query, engine)
options_df = pd.read_sql(options_query, engine)
departures_df = pd.read_sql(departures_query, engine)
```

```python
dataframes = {
    "tours": tours_df,
    "tour_options": tour_options_df,
    "itineraries": itineraries_df,
    "locations_visited": locations_visited_df,
    "options": options_df,
    "departures": departures_df,
}
```

```python
directory = "../../data/processed"
```

```python
for name, df in dataframes.items():
    file_path = os.path.join(directory, f"{name}.csv")
    df.to_csv(file_path, index=False)
```
