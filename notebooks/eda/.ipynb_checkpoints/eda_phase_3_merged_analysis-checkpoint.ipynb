```python
file_path = '../../data/processed/merged.csv'
```

```python
data = pd.read_csv(file_path)
```

```python
data
```

```python
# Extract text from specified columns
columns_to_merge = ["tour_name", "tour_option_name", "description", "travelInclusions", "diningInclusions"] + [f"day_{i}_title" for i in range(1, 29)] + [f"day_{i}_description" for i in range(1, 29)]
merged_texts = data[columns_to_merge].fillna("").apply(lambda row: " ".join(row), axis=1)

# Calculate the length of each merged text and find the maximum length
max_length = merged_texts.apply(len).max()
print(max_length)
```

```python
text_data = data[columns_to_merge]
```

```python
text_data
```

```python
directory = "../../data/processed"
```

```python
file_path = os.path.join(directory, f"text_data.csv")
text_data.to_csv(file_path, index=False)
```
