```python
def print_json_tree(json_obj, level=0, prefix=''):
    if isinstance(json_obj, dict):
        for key in json_obj.keys():
            print('  ' * level + prefix + str(key))
            print_json_tree(json_obj[key], level + 1, key + ' -> ')
    elif isinstance(json_obj, list):
        if json_obj:
            print('  ' * level + prefix + 'List [')
            print_json_tree(json_obj[0], level + 1, prefix + '0 -> ')
            print('  ' * level + ']')
    else:
        print('  ' * level + prefix + str(type(json_obj)))
```

```python
with open('tour-simplified.json', 'r') as file:
    data = json.load(file)
    
print_json_tree(data)
```
