```python
import pandas as pd
import os
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
directory1 = "../text_vectorization/embeddings"
```

```python
file_path = os.path.join(directory1, f"all_embeddings_merged.npy")
merged_embeddings = np.load(file_path)
```

```python
merged_embeddings
```

```python
directory2 = "../../data/processed"
```

```python
encoded_scaled = pd.read_csv(os.path.join(directory2,"categoricals_numericals_encoded_scaled.csv"))
```

```python
encoded_scaled
```

```python
embeddings_df = pd.DataFrame(merged_embeddings, columns=[f'embedding_{i}' for i in range(merged_embeddings.shape[1])])
```

```python
# embeddings_df
```

```python
final_df = pd.concat([encoded_scaled, embeddings_df], axis=1)
```

```python
for col in final_df.columns:
    if final_df[col].dtype == 'bool':
        final_df[col] = final_df[col].astype(int)
```

```python
# Convert the DataFrame to a numpy array
final_array = final_df.values

file_path = os.path.join(directory2, f"final_array.npy")

# Save the numpy array to a file
np.save(file_path, final_array)

file_path = os.path.join(directory2, f"final_df.csv")

final_df.to_csv(file_path, index=False)
```
