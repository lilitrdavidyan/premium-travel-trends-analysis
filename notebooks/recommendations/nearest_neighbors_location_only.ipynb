```python
from sklearn.neighbors import NearestNeighbors
import pandas as pd
import os
import numpy as np
import pickle
```

```python
directory = "../../data/processed"
```

```python
pd.set_option('display.max_columns', None)
```

```python
file_path = os.path.join(directory, f"final_array.npy")
final = np.load(file_path)
```

```python
len(final)
```

```python
len(final[0])
```

```python
# Step 1: Fit the model
model_knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=4, n_jobs=-1)
model_knn.fit(final)
```

```python
# Step 2: Choose a row index
index = 10
```

```python
# Step 3: Find the nearest neighbors
distances, indices = model_knn.kneighbors(final[index].reshape(1, -1), n_neighbors=6)
```

```python
# Step 4: Extract indices and distances of the neighbors
neighbor_indices = indices[0][1:]
neighbor_distances = distances[0][1:]
```

```python
original_data = pd.read_csv(file_path)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute the cosine similarity between tours
similarity_matrix = cosine_similarity(final)

# Get the top-k similar tours for a given tour
tour_index = 10  # Index of the given tour
top_k = 5  # Number of similar tours to retrieve
similar_tours = similarity_matrix[tour_index].argsort()[::-1][:top_k + 1][1:]
```
