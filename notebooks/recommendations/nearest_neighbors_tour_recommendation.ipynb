```python
from sklearn.neighbors import NearestNeighbors
import pandas as pd
import os
import numpy as np
import pickle
```

```python
directory = "../text_vectorization/embeddings"
```

```python
pd.set_option('display.max_columns', None)
```

```python
file_path = os.path.join(directory, f"all_embeddings_merged.npy")
merged_embeddings = np.load(file_path)
```

```python
len(merged_embeddings)
```

```python
len(merged_embeddings[0])
```

```python
# Step 1: Fit the model
model_knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=4, n_jobs=-1)
model_knn.fit(merged_embeddings)
```

```python
# Step 2: Choose a row index
index = 10
```

```python
# Step 3: Find the nearest neighbors
distances, indices = model_knn.kneighbors(merged_embeddings[index].reshape(1, -1), n_neighbors=4)
```

```python
# Step 4: Extract indices and distances of the neighbors
neighbor_indices = indices[0][1:]
neighbor_distances = distances[0][1:]
```

```python
file_path = '../../data/processed/merged.csv'
```

```python
original_data = pd.read_csv(file_path)
```

```python
# Step 5: Extract the rows corresponding to the nearest neighbors
nearest_rows = original_data.iloc[np.append(neighbor_indices,index) ]
```
